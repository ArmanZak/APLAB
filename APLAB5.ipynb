```python
class Holiday:
    def __init__(self, name, day, month):
        self.name = name
        self.day = day
        self.month = month

    def inSameMonth(self, other):
        return self.month == other.month

    @staticmethod
    def avgDate(holidays):
        total = 0
        for h in holidays:
            total += h.day
        return total / len(holidays)


h1 = Holiday("Independence Day", 15, "August")
```

```python
class Movie:
    def __init__(self, title, studio, rating="PG"):
        self.title = title
        self.studio = studio
        self.rating = rating

    @staticmethod
    def getPG(movies):
        pg_movies = []
        for movie in movies:
            if movie.rating == "PG":
                pg_movies.append(movie)
        return pg_movies


m1 = Movie("Casino Royale", "Eon Productions", "PG-13")
```
