```python
alphabet = input("Enter an alphabet: ")

if 'a' <= alphabet <= 'z':
    print(alphabet.upper())
elif 'A' <= alphabet <= 'Z':
    print(alphabet.lower())
else:
    print("Invalid input. Please enter a single alphabet.")
```

```python
n_terms = int(input("Enter the number of terms for Fibonacci series: "))

a, b = 0, 1
count = 0

if n_terms <= 0:
    print("Please enter a positive integer.")
elif n_terms == 1:
    print("Fibonacci series up to 1 term:", a)
else:
    print("Fibonacci series:")
    while count < n_terms:
        print(a, end=" ")
        nth = a + b
        a = b
        b = nth
        count += 1
    print()
```

```python
num = int(input("Enter a number to find its distinct prime factors: "))
factors = []
d = 2

while d * d <= num:
    if num % d == 0:
        if d not in factors:
            factors.append(d)
        num //= d
    else:
        d += 1

if num > 1 and num not in factors:
    factors.append(num)

if not factors:
    print("No prime factors (or input is 0 or 1).")
else:
    print("Distinct prime factors:", end=" ")
    for i in range(len(factors)):
        print(factors[i], end="")
        if i < len(factors) - 1:
            print(",", end=" ")
    print()
```

```python
decimal_num = int(input("Enter a decimal number: "))

print("Choose conversion type:")
print("1) Binary")
print("2) Octal")
print("3) Hexadecimal")

choice = int(input("Enter your choice (1, 2, or 3): "))

if choice == 1:
    print("Binary:", bin(decimal_num))
elif choice == 2:
    print("Octal:", oct(decimal_num))
elif choice == 3:
    print("Hexadecimal:", hex(decimal_num))
else:
    print("Invalid choice.")
```

```python
import random
secret_number = random.randint(1, 100)
guess = 0
guess_count = 0
max_guesses = 7 

print("Welcome to the Guessing Game!")
print(f"I'm thinking of a number between 1 and 100. You have {max_guesses} guesses.")

while guess != secret_number and guess_count < max_guesses:
    try:
        guess = int(input(f"Guess #{guess_count + 1}: Enter your guess: "))
        guess_count += 1

        if guess < secret_number:
            print("Too low!")
        elif guess > secret_number:
            print("Too high!")
        else:
            print(f"Congratulations! You guessed the number {secret_number} in {guess_count} guesses!")
            break 
    except ValueError:
        print("Invalid input. Please enter a whole number.")

if guess != secret_number:
    print(f"\nSorry, you ran out of guesses! The secret number was {secret_number}.")

print("Thanks for playing!")
```
