```python
my_list = [10, 50, 30, 20, 40]

num_items = len(my_list)
print(f"Number of items in the list: {num_items}")

largest_number = max(my_list)
print(f"Largest number in the list: {largest_number}")

smallest_number = min(my_list)
print(f"Smallest number in the list: {smallest_number}")

sum_of_items = sum(my_list)
print(f"Sum of all items in the list: {sum_of_items}")
```

```python
my_list = [10, 25, 5, 80, 15, 30]

target_number = int(input("Enter a number from the given list:"))

found = False

for i in range(len(my_list)):
    
    if my_list[i] == target_number:
        print(f"The number {target_number} is found at position {i}.")
        found = True
        break  

if not found:
    print(f"The number {target_number} is not in the list.")
```

```python
t=(1,2,3,4,5)
rt=t[::-1]
x=int(input("Enter element to remove from reversed tuple: "))
rt=list(rt)
if x in rt: rt.remove(x)
rt=tuple(rt)+(10,20,30)
print("Final tuple:",rt)
```

```python
t=(1,2,3,2,4,2,5)
x=int(input("Enter element to find frequency: "))
print("Frequency:",t.count(x))
print("Index positions:",[i for i,v in enumerate(t) if v==x])
```

```python
lst=[1,2,2,3,4,1,2,3,4,5]
best=[]; cur=[lst[0]]
for i in range(1,len(lst)):
    if lst[i]>lst[i-1]:
        cur.append(lst[i])
    else:
        if len(cur)>len(best): best=cur
        cur=[lst[i]]
if len(cur)>len(best): best=cur
print("Length:",len(best))
print("Subsequence:",best)
```
