```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("company_sales_data_matplotlib.csv")
```

```python
monthList = df['month_number'].tolist()
totalProfit = df['total_profit'].tolist()

plt.figure()
plt.plot(monthList, totalProfit)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit Per Month')
plt.show()
```

```python
plt.figure()
plt.plot(monthList, totalProfit,
         linestyle='dotted',
         color='red',
         marker='o',
         markerfacecolor='red',
         linewidth=3,
         label='Profit Data')

plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.figure()

plt.plot(monthList, df['facecream'], label='Face Cream')
plt.plot(monthList, df['facewash'], label='Face Wash')
plt.plot(monthList, df['toothpaste'], label='Toothpaste')
plt.plot(monthList, df['bathingsoap'], label='Bathing Soap')
plt.plot(monthList, df['shampoo'], label='Shampoo')
plt.plot(monthList, df['moisturizer'], label='Moisturizer')

plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.scatter(monthList, df['toothpaste'])
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales Data')
plt.show()
```

```python
plt.figure()
width = 0.3
monthList_np = np.array(monthList)

plt.bar(monthList_np - width/2, df['facecream'], width=width, label='Face Cream')
plt.bar(monthList_np + width/2, df['facewash'], width=width, label='Face Wash')

plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.bar(monthList, df['bathingsoap'])
plt.xlabel('Month Number')
plt.ylabel('Bathing Soap Sales')
plt.title('Bathing Soap Sales Data')

plt.savefig('bathing_soap_sales.png')
plt.show()
```

```python
plt.figure()
plt.hist(totalProfit, bins=6)
plt.xlabel('Profit Range')
plt.ylabel('Frequency')
plt.title('Profit Distribution')
plt.show()
```

```python
salesData = [
    df['facecream'].sum(),
    df['facewash'].sum(),
    df['toothpaste'].sum(),
    df['bathingsoap'].sum(),
    df['shampoo'].sum(),
    df['moisturizer'].sum()
]

labels = ['Face Cream', 'Face Wash', 'Toothpaste',
          'Bathing Soap', 'Shampoo', 'Moisturizer']

plt.figure()
plt.pie(salesData, labels=labels, autopct='%1.1f%%')
plt.title('Total Sales per Product')
plt.show()
```

```python
plt.figure()

plt.subplot(2, 1, 1)
plt.plot(monthList, df['bathingsoap'])
plt.title('Bathing Soap Sales')

plt.subplot(2, 1, 2)
plt.plot(monthList, df['facewash'])
plt.title('Facewash Sales')

plt.show()
```

```python
plt.figure()

plt.stackplot(monthList,
              df['facecream'],
              df['facewash'],
              df['toothpaste'],
              df['bathingsoap'],
              df['shampoo'],
              df['moisturizer'],
              labels=['Face Cream', 'Face Wash', 'Toothpaste',
                      'Bathing Soap', 'Shampoo', 'Moisturizer'])

plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend(loc='upper left')
plt.show()
```
