```python
# ===============================
# IMPORT LIBRARIES
# ===============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from scipy import stats

# ---- Classification Models ----
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (RandomForestClassifier, GradientBoostingClassifier,
                               AdaBoostClassifier, ExtraTreesClassifier)
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                              f1_score, classification_report,
                              confusion_matrix, ConfusionMatrixDisplay)

sns.set_style("whitegrid")

# ===============================
# Q1: LOAD DATASET & BASIC INFO
# ===============================
print("\n========== Q1: DATA LOADING ==========\n")
df = pd.read_csv('advertising  (2).csv')
print("First 5 Rows:\n", df.head())
print("\nDataset Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())
print("\nDataset Info:"); df.info()

# ===============================
# Q2: MISSING VALUE TREATMENT
# ===============================
print("\n========== Q2: MISSING VALUE DETECTION ==========\n")
print("Missing Values:\n", df.isnull().sum())
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
df.dropna(inplace=True)
print("\nAfter Treatment:\n", df.isnull().sum())

# ===============================
# Q3 & Q4: ENCODING
# ===============================
print("\n========== Q4: ENCODING CATEGORICAL VARIABLES ==========\n")
cols_to_drop = ['Name', 'Ticket', 'Cabin']
df.drop(columns=[c for c in cols_to_drop if c in df.columns], inplace=True)
cat_cols = df.select_dtypes(include='object').columns
if len(cat_cols) > 0:
    le = LabelEncoder()
    for col in cat_cols:
        df[col] = le.fit_transform(df[col])
        print(f"Label encoded: {col}")

X = df.iloc[:, :-1]
y = df.iloc[:, -1]
print("\nX shape:", X.shape, "| y shape:", y.shape)

# ===============================
# Q5: TRAIN TEST SPLIT
# ===============================
print("\n========== Q5: TRAIN TEST SPLIT ==========\n")
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print("Train size:", X_train.shape, "| Test size:", X_test.shape)

# ===============================
# Q6: FEATURE SCALING
# ===============================
print("\n========== Q6: FEATURE SCALING ==========\n")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# ===============================
# Q7: UNIVARIATE ANALYSIS
# ===============================
print("\n========== Q7: UNIVARIATE ANALYSIS ==========\n")
col = X.columns[0]
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(X[col], kde=True, ax=axes[0]); axes[0].set_title(f"Histogram - {col}")
sns.boxplot(y=X[col], ax=axes[1]);          axes[1].set_title(f"Boxplot - {col}")
plt.tight_layout(); plt.show()
print("Skewness:", X[col].skew())

# ===============================
# Q8: BIVARIATE ANALYSIS
# ===============================
print("\n========== Q8: BIVARIATE ANALYSIS ==========\n")
num_c = X.select_dtypes(include=np.number).columns
if len(num_c) >= 2:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=X[num_c[0]], y=X[num_c[1]])
    plt.title("Scatter Plot"); plt.show()
    print("Pearson Correlation:", X[num_c[0]].corr(X[num_c[1]]))

# ===============================
# Q9: MULTIVARIATE ANALYSIS
# ===============================
print("\n========== Q9: MULTIVARIATE ANALYSIS ==========\n")
plt.figure(figsize=(8, 6))
sns.heatmap(X.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap"); plt.show()

# ===============================
# Q10: OUTLIER DETECTION
# ===============================
print("\n========== Q10: OUTLIER DETECTION ==========\n")
z = np.abs(stats.zscore(X.select_dtypes(include=np.number)))
iso = IsolationForest(contamination=0.05, random_state=0)
outliers_iso = iso.fit_predict(X)
lof = LocalOutlierFactor(n_neighbors=20)
outliers_lof = lof.fit_predict(X)
print("Isolation Forest outliers detected:", (outliers_iso == -1).sum())
print("LOF outliers detected:", (outliers_lof == -1).sum())


# ================================================================
# Q11: 10 CLASSIFICATION MODELS — FULL EVALUATION
# ================================================================

print("\n========== Q11: CLASSIFICATION MODEL TRAINING & EVALUATION ==========\n")

# Determine averaging strategy
num_classes = y.nunique()
avg = 'binary' if num_classes == 2 else 'weighted'
print(f"Target classes: {num_classes}  |  Averaging strategy: '{avg}'\n")

# ------------------------------------------------------------------
# Define all 10 models
# ------------------------------------------------------------------
models = {
    "Logistic Regression"         : LogisticRegression(max_iter=1000, random_state=0),
    "Decision Tree"               : DecisionTreeClassifier(random_state=0),
    "Random Forest"               : RandomForestClassifier(n_estimators=100, random_state=0),
    "Support Vector Machine"      : SVC(kernel='rbf', probability=True, random_state=0),
    "K-Nearest Neighbors"         : KNeighborsClassifier(n_neighbors=5),
    "Naive Bayes"                 : GaussianNB(),
    "Gradient Boosting"           : GradientBoostingClassifier(n_estimators=100, random_state=0),
    "AdaBoost"                    : AdaBoostClassifier(n_estimators=100, random_state=0),
    "Linear Discriminant Analysis": LinearDiscriminantAnalysis(),
    "Extra Trees"                 : ExtraTreesClassifier(n_estimators=100, random_state=0),
}

# ------------------------------------------------------------------
# Storage for comparison
# ------------------------------------------------------------------
results = {
    "Model"    : [],
    "Accuracy" : [],
    "Precision": [],
    "Recall"   : [],
    "F1 Score" : [],
}

# ------------------------------------------------------------------
# Train + Evaluate each model
# ------------------------------------------------------------------
for name, model in models.items():
    print(f"\n{'='*60}")
    print(f"  MODEL: {name}")
    print(f"{'='*60}")

    # Train
    model.fit(X_train_scaled, y_train)

    # Predict
    y_pred = model.predict(X_test_scaled)

    # Metrics
    acc  = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average=avg, zero_division=0)
    rec  = recall_score(y_test, y_pred, average=avg, zero_division=0)
    f1   = f1_score(y_test, y_pred, average=avg, zero_division=0)

    # Print individual metrics
    print(f"\n  Accuracy  : {acc:.4f}  ({acc*100:.2f}%)")
    print(f"  Precision : {prec:.4f}  ({prec*100:.2f}%)")
    print(f"  Recall    : {rec:.4f}  ({rec*100:.2f}%)")
    print(f"  F1 Score  : {f1:.4f}  ({f1*100:.2f}%)")

    # Full classification report
    print(f"\n  Classification Report:\n")
    print(classification_report(y_test, y_pred, zero_division=0))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    plt.figure(figsize=(5, 4))
    disp.plot(cmap='Blues', colorbar=False)
    plt.title(f"Confusion Matrix — {name}")
    plt.tight_layout()
    plt.show()

    # Save to results
    results["Model"].append(name)
    results["Accuracy"].append(acc)
    results["Precision"].append(prec)
    results["Recall"].append(rec)
    results["F1 Score"].append(f1)


# ------------------------------------------------------------------
# SUMMARY TABLE
# ------------------------------------------------------------------
print("\n========== SUMMARY TABLE: ALL MODELS ==========\n")
summary_df = pd.DataFrame(results)
summary_df = summary_df.sort_values("Accuracy", ascending=False).reset_index(drop=True)
summary_df.index += 1

# Format as percentages for display
display_df = summary_df.copy()
for col in ["Accuracy", "Precision", "Recall", "F1 Score"]:
    display_df[col] = display_df[col].apply(lambda x: f"{x*100:.2f}%")

print(display_df.to_string())
best = summary_df.iloc[0]
print(f"\n✅ Best Model : {best['Model']}")
print(f"   Accuracy  : {best['Accuracy']*100:.2f}%")
print(f"   Precision : {best['Precision']*100:.2f}%")
print(f"   Recall    : {best['Recall']*100:.2f}%")
print(f"   F1 Score  : {best['F1 Score']*100:.2f}%")


# ------------------------------------------------------------------
# COMPARISON CHARTS — All 4 Metrics
# ------------------------------------------------------------------
metrics   = ["Accuracy", "Precision", "Recall", "F1 Score"]
colors    = ["steelblue", "seagreen", "tomato", "darkorange"]
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle("Model Comparison — All Metrics", fontsize=15, fontweight='bold')

for ax, metric, color in zip(axes.flatten(), metrics, colors):
    vals   = summary_df[metric].values * 100
    names  = summary_df["Model"].values
    bars   = ax.barh(names, vals, color=color, edgecolor='white')
    ax.set_xlim(0, 110)
    ax.set_xlabel("Score (%)")
    ax.set_title(metric, fontweight='bold')
    for bar, v in zip(bars, vals):
        ax.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height() / 2,
                f"{v:.2f}%", va='center', fontsize=8)

plt.tight_layout()
plt.show()


# ------------------------------------------------------------------
# GROUPED BAR CHART — Side-by-side comparison
# ------------------------------------------------------------------
x      = np.arange(len(summary_df["Model"]))
width  = 0.2
fig, ax = plt.subplots(figsize=(16, 6))

for i, (metric, color) in enumerate(zip(metrics, colors)):
    ax.bar(x + i * width, summary_df[metric] * 100, width,
           label=metric, color=color, edgecolor='white')

ax.set_xticks(x + width * 1.5)
ax.set_xticklabels(summary_df["Model"], rotation=25, ha='right', fontsize=9)
ax.set_ylabel("Score (%)")
ax.set_title("All Models — Accuracy, Precision, Recall & F1 Score", fontweight='bold')
ax.legend(loc='lower right')
ax.set_ylim(0, 115)
plt.tight_layout()
plt.show()
```
