```python
class Bigmart:
    total_amount = 0

    def __init__(self, serial_number, name, amount_spend):
        self.serial_number = serial_number
        self.name = name
        self.amount_spend = amount_spend

        Bigmart.total_amount += self.amount_spend

    def Display(self):
        print(f"Serial No : {self.serial_number}")
        print(f"Name      : {self.name}")
        print(f"Amount    : ₹{self.amount_spend}")

c1 = Bigmart(1, "Alice",   1500)
c2 = Bigmart(2, "Bob",     2300)
c3 = Bigmart(3, "Charlie", 800)
c4 = Bigmart(4, "Diana",   3200)

print("BIGMART - Daily Sales Report")
c1.Display()
c2.Display()
c3.Display()
c4.Display()

print(f"Total Amount Sold Today: ₹{Bigmart.total_amount}")
```
